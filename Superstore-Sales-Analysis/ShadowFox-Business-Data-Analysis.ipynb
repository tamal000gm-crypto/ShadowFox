To analyze sales and profit trends from the Superstore dataset using Python.

## Step 1: Import Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Step 2: Upload Dataset

```python
from google.colab import files
uploaded = files.upload()
```

## Step 3: Load and Preview Dataset

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
df.head()
```

## Step 4: Dataset Information and Summary

```python
df.info()
```

```python
df.describe()
```

## Step 5: Checking Missing Values

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

## Step 6: Total Sales and Profit

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

## Step 7: Sales by Category

```python
plt.figure()
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

## Step 8: Profit by Category

```python
plt.figure()
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

## Step 9: Monthly Sales Trend

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.month

plt.figure()
df.groupby('Month')['Sales'].sum().plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

## Conclusion

The Superstore dataset analysis provides insights into overall business performance.
Technology category shows strong sales performance.
Profit distribution varies across categories and states.
Monthly trends help understand seasonal demand.
